# scale the features in preparation for ML

```python
features_list=df.columns[1:-1]
features_list
```

```python
#before scaling, check the spread/outliers
df[features_list].boxplot();
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[features_list]=scaler.fit_transform(df[features_list])

df.describe()
```

```python
df[features_list].boxplot();
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, max_error
import math

#use the scaled data
X=df[features_list]
y=df['Store_Sales']
Xtrain,Xtest,ytrain, ytest = train_test_split(X,y,test_size=0.3)

model=LinearRegression()

cv_score = cross_validate(model, Xtrain,ytrain,cv=10, \
    scoring=['max_error','r2','neg_mean_absolute_error','neg_mean_squared_error'])

```

```python
cv_score
```

```python
for key, value in cv_score.items():
    print(f"{key} {np.mean(value)}")
```

```python
model.fit(Xtrain,ytrain)
y_pred=model.predict(Xtest)

mse=mean_squared_error(ytest,y_pred)
rmse=math.sqrt(mse)
mae=mean_absolute_error(ytest,y_pred)
Rsquared = r2_score(ytest,y_pred)
Max_err = max_error(ytest,y_pred)

print(f"mse {mse}, rmse {rmse}, mae {mae}, Rsquared {Rsquared}, Max err {Max_err}")
```

```python
#try solvers
from sklearn.linear_model import SGDRegressor

sql_sgd= SGDRegressor(loss="squared_loss")
cv_score_sgd = cross_validate(sql_sgd, Xtrain,ytrain,cv=10, \
    scoring=['max_error','r2','neg_mean_absolute_error','neg_mean_squared_error'])


cv_score_sgd
```

```python
for key, value in cv_score_sgd.items():
    print(f"{key} {np.mean(value)}")
```

```python
#Solver = MAE
mae_sgd= SGDRegressor(loss="epsilon_insensitive", epsilon = 0)

cv_mae_sgd = cross_validate(mae_sgd, Xtrain,ytrain,cv=10, \
    scoring=['max_error','r2','neg_mean_absolute_error','neg_mean_squared_error'])


cv_mae_sgd
```

```python
for key, value in cv_mae_sgd.items():
    print(f"{key} {np.mean(value)}")
```

```python
#KNN
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
#find best n-neighbors for the KNN model
best_k=[]
for knn in range(2,20):
    neigh = KNeighborsRegressor(n_neighbors=knn)
    neigh.fit(Xtrain, ytrain)
    score=cross_val_score(neigh,Xtrain, ytrain, cv=5)
    best_k.append(score.mean())
```

```python
best_k
```

```python
plt.plot(best_k);
plt.xlabel('knn')
plt.ylabel('cross_val_score')
```

```python
best_score=max(best_k)
best_k=best_k.index(best_score)
best_k
```

```python
neigh = KNeighborsRegressor(n_neighbors=17)
neigh.fit(Xtrain, ytrain)
y_pred=neigh.predict(Xtest)

mse=mean_squared_error(ytest,y_pred)
rmse=math.sqrt(mse)
mae=mean_absolute_error(ytest,y_pred)
Rsquared = r2_score(ytest,y_pred)
Max_err = max_error(ytest,y_pred)

print(f"mse {mse}, rmse {rmse}, mae {mae}, Rsquared {Rsquared}, Max err {Max_err}")
```

Third Conclusion:
LinearRegression model predicted with slightly less error as KNN, better than SGDRegressor
